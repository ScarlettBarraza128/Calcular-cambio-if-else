### Las variables
1. CP --> Costo del producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$

## Diagrama de flujo LucidChart

<img src=https://raw.githubusercontent.com/ScarlettBarraza128/Calcular-cambio-if-else/refs/heads/main/DF%20if%20else%20Cambio.png alt="Imagen de ejemplo" width="400" height="550">

```python

# para cuando es Verdadero (SI)
# Calculando el cambio usando if else
# Datos de entrada CD,  CP,
# Mostrar Cambio si es falso o
# Si es verdadero mostrar no alcanza
CD = float(input("Ingrese la cantidad de dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("Cambio: ", Cambio)
    print("Fin del programa")
    print("Programa realizado por Emily Barraza NL = 7")
```

```python
# para cuando es Falso (NO)
# Calculando el cambio usando if else
# Datos de entrada CD,  CP,
# Mostrar Cambio si es falso o
# Si es verdadero mostrar no alcanza
CD = float(input("Ingrese la cantidad de dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP > CD:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("Cambio: ", Cambio)
    print("Fin del programa")
    print("Programa realizado por Emily Barraza NL = 7")
```
